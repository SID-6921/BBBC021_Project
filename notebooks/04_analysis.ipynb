# 04 - Analysis
Compare metrics between Group A and Group B and generate plots.

```python
from pathlib import Path
import sys
import pandas as pd
import matplotlib.pyplot as plt

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
SRC_DIR = PROJECT_ROOT / 'src'
if str(SRC_DIR) not in sys.path:
    sys.path.append(str(SRC_DIR))

from visualize import save_group_plots

metrics_path = PROJECT_ROOT / 'outputs' / 'metrics' / 'image_metrics.csv'
plots_dir = PROJECT_ROOT / 'outputs' / 'plots'
plots_dir.mkdir(parents=True, exist_ok=True)

df = pd.read_csv(metrics_path)
display(df.head())
print('Rows:', len(df))
```

```python
summary = df.groupby('group')[['spot_count', 'avg_brightness', 'total_intensity', 'area_covered_px', 'area_covered_ratio']].agg(['mean', 'std', 'median'])
display(summary)
summary.to_csv(PROJECT_ROOT / 'outputs' / 'metrics' / 'group_summary.csv')
```

```python
save_group_plots(df, plots_dir, group_col='group')
print('Saved plots to:', plots_dir)

sample_plot = next(plots_dir.glob('*.png'), None)
if sample_plot is not None:
    img = plt.imread(sample_plot)
    plt.figure(figsize=(8, 5))
    plt.imshow(img)
    plt.axis('off')
    plt.title(sample_plot.name)
    plt.show()
```
